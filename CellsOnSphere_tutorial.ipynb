# Initate embryo and run simulation, saving images of embryo visualisation

```python
import CellsOnSphere as CoS

save1 = r'put save path and name here'

# steps of 50 seconds = 3um steps
dT = 50 
# total 4 hours time
endTime = 60*60*4 

cellCycleTime = 60*60*2

E = CoS.Sphere(radius=330,epiboly=0.505,epibolyV=0.004)

E.initiateRing(N=150,r=6.6,v=0.06,p=200,rep=True,cT=cellCycleTime)

#meas = []
for i in range(int(endTime/dT)):
    E.evolveOneStep(dT)
    if i%10==0:
        print(i)
        E.plotImage(resFac=3,save=save1)
        #meas.append(E.measureHomogeneity())
for i in range(int(60*60*4/dT)):
    E.evolveReaggregateStep(dT)
    if i%10==0:
        print(i)
        E.plotImage(resFac=3,save=save1)
        #meas.append(E.measureHomogeneity())
```

# Run another simulation saving homogeneity measurements

```python
import CellsOnSphere as CoS

save1 = r'put save path and name here'

# steps of 50 seconds = 3um steps
dT = 50 
# total 4 hours time
endTime = 60*60*4 

cellCycleTime = 60*60*2

E = CoS.Sphere(radius=330,epiboly=0.505,epibolyV=0.004)

E.initiateRing(N=150,r=6.6,v=0.06,p=300,rep=True,cT=cellCycleTime)
```

```python
meas = []
for i in range(int(endTime/dT)):
    E.evolveOneStep(dT)
    if i%10==0:
        print(i)
        #E.plotImage(resFac=3,save=save1)
        meas.append(E.measureHomogeneity())
```

# Plot homogeneity mesurements

```python
import matplotlib.pyplot as plt
plt.plot(meas)
```

# Another example with different parameters (see docs)

```python
import CellsOnSphere as CoS

save1 = r'put save path and name here'
save2 = r'put save path and name here'
endTime = 3000
dT = 8

E = CoS.Sphere()
E.initiateRing(N=25,r=100,p=300,rep=True)



for i in range(int(endTime/dT)):
    E.evolveOneStep(dT)
    if i%5==0:
        print(i)
        E.plotImage(resFac=1,save=save1)

E = CoS.Sphere()
E.initiateRing(N=25,r=100,p=300,rep=False)
for i in range(int(endTime/dT)):
    E.evolveOneStep(dT)
    if i%5==0:
        print(i)
        E.plotImage(resFac=1,save=save2)    
```
